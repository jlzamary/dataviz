### Bar chart for final project 

**About the data:**

- 14,782 employed adults under the age of 75 across 31 countries
- interviewed May 26 – June 9, 2023 -- online only in all countries except India.

```python
# Import Libraries 
import pandas as pd
import seaborn as sns 
import geopandas as gpd
import numpy as np
import folium 
import geojson 
import json
import matplotlib.pyplot as plt

# Plotly
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Import data
df = pd.read_csv("Data/fig_8.1.7.csv")

# Show data 
df.head()
```

```python
# Data engineering 
# Total respondents surveyed
total_respondents = 14782

# Convert % of respondents to count
df['Count of People Surveyed'] = (
    df['% of respondents'].str.rstrip('%').astype(float) / 100 * total_respondents).round().astype(int)

# Show data
df.head(10)
```

```python
# Plotly plot - Two donut charts for AI impact questions

# Filter data for each question
df_change = df[df['Global opinions on the impact of AI on current jobs'] == 'AI will change how you do your current job in the next 5 years']
df_replace = df[df['Global opinions on the impact of AI on current jobs'] == 'AI will replace your current job in the next 5 years']

# Titles
titles = [
    "<b>AI will change how you do your<br>current job in the next 5 years</b><br>n = 14,782</br>",
    "<b>AI will replace your current<br>job in the next 5 years</b><br>n = 14,782</br>"]

# Make subplots  
fig = make_subplots(
    rows=1, cols=2,
    # From plotly documentation 
    specs=[[{'type': 'domain'}, {'type': 'domain'}]],
    subplot_titles=titles)

# Colors for responses
# From Coolors:
# https://coolors.co/palette/f94144-f3722c-f8961e-f9844a-f9c74f-90be6d-43aa8b-4d908e-577590-277da1
colors = ['#277da1', '#577590', '#4d908e', '#43aa8b', 'lightgrey']

# Response labels
labels = df_change['Response'].tolist()

# Add first donut chart (change job)
fig.add_trace(
    go.Pie(
        labels=labels, 
        values=df_change['Count of People Surveyed'].tolist(), 
        hole=.5, 
        marker_colors=colors,
        name='Change Job',
        texttemplate="%{percent:.1%}"
    ),
    row=1, col=1
)

# Add second donut chart (replace job)
fig.add_trace(
    go.Pie(
        labels=labels, 
        values=df_replace['Count of People Surveyed'].tolist(), 
        hole=.5, 
        marker_colors=colors,
        name='Replace Job',
        texttemplate="%{percent:.0%}"
    ),
    row=1, col=2
)

# Update layout 
fig.update_layout(
    height=500, 
    width=1000, 
    showlegend=True,
)

# Update traces
fig.update_traces(
    textinfo='percent',
    textfont_size=14,
    hovertemplate='<span style="background-color:grey;color:black;font-family:Georgia;font-weight:bold;font-size:16px;"> <b>%{label}</b><br>Count: %{value}<br><extra></extra>', 
    hoverlabel={'bgcolor':'lightgrey'},
    marker={'line':{'color':'#ffffff', 'width':2}}
)

# Show
fig.show()
```
